# Graficar funciones en python

```python
# Importamos las librerías requeridas
import numpy as np
import matplotlib.pyplot as plt
```

## Por qué usar Numpy

Para graficar se puede obviar el uso de numpy y trabajar, por ejemplo, con listas, pero se vuelve más tedioso. Miremoslo con el siguiente ejemplo:

Supongamos que queremos graficar la función $y = 4x + 5$

```python
# Trabajando con Listas
x = [1,2,3,4,5]
"""
Un error sería hacer lo siguiente:
y = 4 * x + 5, ya que esto no va a multiplicar
4 por cada entrada de x y sumarle 5 a cada entrada de x

Hay muchas maneras de lograr lo anteriomente mencionado, pero
para demostrar lo tedioso que puede ser trabajar con listas
se escogió el siguiente método.
"""
# Se crea la variable donde se almacenarán los valores de la función
y = []
# Se recorre cada uno de los valores de x
for elemento in x:
  # A cada elemento se le aplica la función y se almacena en y
  y.append(4*elemento + 5)

# Se demuestra que el código logra los outputs esperados
print(y)

# Para el que aún tenga dudas, eliminar los siguientes comentarios
# y = 4 * x + 5
# print(y)
```

```python
# Trabajando con Numpy
x = np.array([1,2,3,4,5])

"""
Con Numpy el trabajo se reduce demasiado utilizando los arrays.
Miremos como se reduce el trabajo
"""
# Se crea la función pedida
y = 4 * x + 5

# Se demuestra que el código logra los outputs esperados
print(y)
```

## Gráfica de una función sencilla

Una vez comprendido lo anterior, vamos a generar las gráficas de funciones para familizarse con *mathplot*

```python
# Genera valores de x para el dominio
x = np.arange(0,10,0.5)

# Calcula los valores de y para cada función
y = x**2
# Genera la gráfica de x contra y
plt.plot(x, y)
# Se etiquetan cada uno de los ejes
plt.xlabel('x')
plt.ylabel('y')
# Se establece el título de la gráfica
plt.title('Gráfica de la función cuadrática f(x) = x^2')
# Esto es opcional, pero es para ver las celdas de fondo
plt.grid(True)
# Se muestran los resultados  de los pasos anteriores
plt.show()
```

## Gráfica de varias funciones en una misma figura

```python
# Genera valores de x para el dominio
x = np.arange(1,5,0.1)

# Calcula los valores de y para cada función
y1 = x**2
y2 = np.log(x)
y3 = x**3

# Grafica las funciones
plt.plot(x, y1, label='f(x) = x^2')
plt.plot(x, y2, label='f(x) = log(x)')
plt.plot(x, y3, label='f(x) = x^3')

# Añade etiquetas y título al gráfico
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráficas de varias funciones matemáticas')
# Muestra la leyenda con los nombres de las funciones
plt.legend()
plt.grid(True)
plt.show()
```

# Gráfico de áreas usando la función fill

La función fill de matplot nos permite rellenar el área que generan poligonos.

```python
# Coordenadas del cuadrado
x = [0, 1, 1, 0, 0]
y = [0, 0, 1, 1, 0]

# Graficar el cuadrado
plt.plot(x, y, color='blue')
plt.fill(x, y, color='blue', alpha=0.3)  # Rellena el área del cuadrado

# Añadir etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de un cuadrado')

# Ajustar los límites del eje para que el cuadrado esté completamente visible
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

# Gráfico de áreas usando la función fill_between

La función fill_between nos permite graficar el area que hay entre dos curvas.

```python
# Genera valores de x para el dominio
x = np.arange(0,2, 0.01)  # Inicio, fin, paso

# Genera las funciones
y1 = np.sin(2 * np.pi * x)
y2 = 0.8 * np.sin(4 * np.pi * x)

# Grafica el área entre y1 y y2
plt.fill_between(x,y1,y2)
# Ajusta el padding de la figura para mejorar la visualización
plt.tight_layout()

plt.title('Gráficas de funciones matemáticas con áreas sombreadas')
# Muestra la figura
plt.show()
```

## Por qué usar el fill between

Para la visualización de las gráficas, habrá momentos en que es una herramienta apropiada ya que es mucho más sencillo usar esta función que el método fill.

Además, si se profundiza en este concepto, la función *fill_between* permite usar el argumento *where*, donde podemos hacer que se grafiquen cosas siempre que se cumpla alguna condición.

Abajo se demuestra el por qué la función *fill_between* facilita la vida al graficar entre curvas.

```python
# Genera valores de x para el dominio
x = np.arange(0,2, 0.01)

# Genera las funciones
y1 = np.sin(2 * np.pi * x)
y2 = 0.8 * np.sin(4 * np.pi * x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.fill(x,y1)
plt.fill(x,y2)

plt.show()
```

# Grafica del área de solución de un problema de programación lineal

Consideremos el siguiente problema de programación lineal.

$Min$ $Z = 60x_1 + 40x_2$

$S.t$

$4x_1 + 4x_2 \geq 10$

$2x_1 + x_2 \geq 4 $

$6x_1 + 2x_2 \leq 12$

$x_1,x_2 \geq 0$

Graficar el área de solución del problema previamente descrito

```python
"""
Primero, graficamos las funciones que restringen
a la función objetivo. Una vez hecho esto, podemos
hacernos una idea de cómo sería el área de solución.
Después, usando la función fill vamos a graficar el
área de solución
"""
# Generamos el dominio de la función
x = np.arange(0, 4)
# Restricciones del problema
res_1 = 2.5 - x
res_2 = 4 - 2 * x
res_3 = 6 - 3 * x

# Generamos las gráficas de las restricciones
plt.plot(x, res_1 , label = '4x1 + 4x2 >= 10')
plt.plot(x, res_2 , label= ' 2x1 + x2 >= 4')
plt.plot(x, res_3 , label = '6x1 + 2x2 <= 12')

# Se restringe el eje para que el gráfico no sea demasiado grande
plt.axis([0, 3, 0, 6])
plt.grid(True)
plt.legend()
plt.show()
```

```python
"""
IMPORTANTE: Antes de correr esta celda, correr la anterior.

Ya que sabemos como luce el área de la solución
podemos graficarla con el método fill. Para esto
tomamos los puntos X,Y. PD: Los puntos son así:
X = [x1,x2,...], Y = [y1,y2,...].
"""
# Generamos las gráficas de las restricciones
plt.plot(x, res_1 , label = '4x1 + 4x2 >= 10')
plt.plot(x, res_2 , label= ' 2x1 + x2 >= 4')
plt.plot(x, res_3 , label = '6x1 + 2x2 <= 12')
# Definimos los puntos del área de solución
x = [0, 1.5, 1.75, 0]
y = [4, 1, 0.75, 6]
# Graficamos el área de solución
plt.fill(x, y, 'paleturquoise', alpha = 0.3)


plt.text(0.1, 4, 'Área \n Solución', size = '11')
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.title('Área de solución de un problema de programación lineal')
plt.axis([0, 3, 0, 6])
plt.grid(True)
plt.legend()
plt.show()
```
